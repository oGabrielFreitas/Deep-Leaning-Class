####ETAPA DE PROCESSAMENTO E ANÁLISE DOS DADOS 

Aqui fazemos a limpeza e processamento do dataset, na ordem: 

                                            -Leitura de todas as linhas
                                            -Remoção do cabeçalho
                                            -Remoção do newline de cada linha
                                            -Adiciona linhas decodificadas a um array (data)
                                            -Printa o datased decodificado sem cabeçalho
                                            -Printa o numero de linhas do dataset
                                            -Criamos um novo array com os dados separados por ','
                                            -Criamos um novo array sem os dados de ID e saida (fdata)
                                            -Criamos um array do numpy com os dados (nparray)
                                            -Criamos o array de saidas no formato one_hot
                                            -Populamos o array one_hot com as saidas
                                            -Printa o array de entradas
                                            -Printa o array de saidas
                                            -Encontra e armazena os valores min/max das entradas
                                            -Normaliza o array entre 0 - 1
                                            -Printa o min/max do array normalizado
                                            -Printa o array normalizado


```
#Retorna o ponteiro de leitura sempre para o inicio do arquivo
f.seek(0) 

#Leitura de todas as linhas 
lines = f.readlines()

#Criação de um array vazio
data = []

#For percorrendo as linhas 
for line in lines[1:]: #[1:] "pula" a primeira linha
  data.append(line.decode()[:-1]) #adiciona a linha decoficada, removendo o \n ([:-1] )

#Printa o dataset decodificado sem o cabeçalho
print(data)

#Printa o numero de linhas do dataset
print(len(data))

#Separação dos dados de cada linha utilizando o delimitador "," 
splitData = [str.split(",") for str in data]

#Gerado um novo dataset, "ignorando" a primeira e a ultima coluna
#Já que a primeira coluna são os ID's e a Ultima são as classes
fData = [x[1:5] for x in splitData]

#Criação de um array do numpay, definindo os dados como ponto flutuante
nparray = np.array(fData,dtype="float")

#Array de zeros, a ser preenchido com as saidas do one-hot
one_hot = np.zeros((len(lines)-1,3))

#Define as possiveis categorias de saida para o one-hot
categ = np.array(['Iris-setosa','Iris-versicolor','Iris-virginica'])

#Cria o array com saidas one-hot
for line in range(len(data)):
  one_hot[line] = (categ == splitData[line][5]).astype('float')

#Printa o array one-hot
print(one_hot)

#Printa o array de entradas
print(nparray)

#Encontrando o maior valor para cada coluna (axis=0)
col0max, col1max, col2max, col3max = nparray.max(axis=0)

#Encontrando o menor valor para cada coluna (axis=0)
col0min, col1min, col2min, col3min = nparray.min(axis=0)
```

```
#Salva os valores de min e max globais
min = nparray.min()
max = nparray.max()

#Criação do array a ser utilizado na normalização
array_norm = np.empty_like(nparray)

#Realização da normalização, salvando o resultado no "array_norm"
# for i in range(len(nparray)):
#   for j in range(len(nparray[i])):
    #norm=(xi−min(x))/(max(x)−min(x))
array_norm= (nparray - min)/(max-min)

#Comprovando que os valores do array array_norm estão entre 0 e 1
print(array_norm.min(), array_norm.max())

#Print do array normalizado
print(array_norm)

```

####ETAPA DE SEPARAÇÃO FINAL

Aqui finalizamos o processamento do dataset, nos seguintes passos:
                                                                
                                                                -Embaralhamos as entradas e saidas, mantendo os pares
                                                                -Separamos em dados de treinamento e validação, na proporçao 90/10


```
#Embaralha entradas e saidas, mantendo os pares correspondentes
mapIndexPosition = list(zip(nparray, one_hot))

np.random.shuffle(mapIndexPosition)
nparray, one_hot = zip(*mapIndexPosition)
```
